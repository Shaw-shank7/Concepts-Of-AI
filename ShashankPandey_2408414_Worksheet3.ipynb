```python
# 3.1 Warm up exercise
# 1. Sorting and Subsetting

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Week2/Dataset/Titanic-Dataset.csv")

# 2. Check the info of the DataFrame and identify the following:
df.info()

# Problem 1 - Sorting:
# 1. Create a DataFrame called 'fare' containing only the 'Fare' column
fare = df[['Fare']]
print("\nFare DataFrame:")
print(fare.head())

# 2. Create a DataFrame called 'class_age' containing 'Pclass' and 'Age' columns in that order
class_age = df[['Pclass', 'Age']]
print("\nClass Age DataFrame:")
print(class_age.head())

# 3. Create a DataFrame called 'survived_gender' containing 'Survived' and 'Sex' columns in that order
survived_gender = df[['Survived', 'Sex']]
print("\nSurvived Gender DataFrame:")
print(survived_gender.head())
```

```python
# Problem 2 - Subsetting:

# 1. Filter the Titanic dataset for cases where the passenger’s fare is greater than 100, assigning it to fare gt 100.View the printed result.
fare_gt_100 = df[df['Fare'] > 100]
print("Passengers with Fare greater than 100:")
print(fare_gt_100)

# 2. Filter the Titanic dataset for cases where the passenger’s class (Pclass) is 1, assigning it to first class. View the printed result.
first_class = df[df['Pclass'] == 1]
print("\nPassengers in PClass 1 (First Class):")
print(first_class)

# 3. Filter the Titanic dataset for cases where the passenger’s age is less than 18 and the passenger is female (Sex is "female"),
# assigning it to female under 18. View the printed result.
female_under_18 = df[(df['Age'] < 18) & (df['Sex'] == 'female')]
print("\nFemale Passengers under 18:")
print(female_under_18)
```

```python
# Subsetting Rows by Categorical variables:

# 1. Filter the Titanic dataset for passengers whose Embarked port is either "C" (Cherbourg) or "S" (Southampton),
# assigning the result to embarked c or s. View the printed result.
embarked_c_or_s = df[df['Embarked'].isin(['C', 'S'])]
print("Passengers who Embarked at Cherbourg (C) or Southampton (S):")
print(embarked_c_or_s)

#2. Filter the Titanic dataset for passengers whose Pclass is in the list [1, 2] (indicating first or second class),
# assigning the result to first second class.View the printed result.
first_second_class = df[df['Pclass'].isin([1, 2])]
print("\nPassengers in First or Second Class:")
print(first_second_class)
```

```python
# 3.2 Exploratory Data Analysis Practice Exercise - 1.
# Warning: Handle missing values in the Age column by filling them with the median age of the dataset before performing the division.)

# Handle missing values in the 'Age' column by filling with the median age
df = df.copy()  # To avoid the SettingWithCopyWarning
df['Age'] = df['Age'].fillna(df['Age'].median())

# 1. Add a column to the Titanic dataset, fare per year, containing the fare divided by the age of the passenger(i.e., Fare/Age).
df['fare_per_year'] = df['Fare'] / df['Age']

# 2. Subset rows where fare per year is higher than 5, assigning this to high fare age.
high_fare_age = df[df['fare_per_year'] > 5]

# 3. Sort high fare age by descending fare per year, assigning this to high fare age srt.
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)

# 4. Select only the Name and fare per year columns of high fare age srt and save the result as result.
result = high_fare_age_srt[['Name', 'fare_per_year']]

# 5. Look at the result.
print("Passengers with the highest fare paid relative to their age:")
print(result)
```

```python
# 3.2 Exploratory Data Analysis Practice Exercise - 2.

# 1. Add a column to the Titanic dataset, fare per class, containing the fare divided by the passenger class i.e., Fare / Pclass.
df['fare_per_class'] = df['Fare'] / df['Pclass']

# 2. Subset rows where the passenger is male (Sex is "male") and an adult (Age is greater than or equal to 18), assigning this to adult males.
adult_males = df[(df['Sex'] == 'male') & (df['Age'] >= 18)]

# 3. Sort adult males by descending fare per class, assigning this to adult males srt.
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

# 4. Select only the Name, Age, and fare per class columns of adult males srt and save the result as result.
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]

# 5. Look at the result.
print("Adult male passenger who paid the highest fare relative to their class:")
print(result)
```

```python
# 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise
# What percent of the total fare revenue came from each passenger class?

# 1. Calculate the total Fare paid across all passengers in the Titanic dataset.
total_fare = df['Fare'].sum()

# 2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.
first_class_fare = df[df['Pclass'] == 1]['Fare'].sum()

# 3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).
second_class_fare = df[df['Pclass'] == 2]['Fare'].sum()
third_class_fare = df[df['Pclass'] == 3]['Fare'].sum()

# 4. Combine the fare totals from first, second, and third classes into a list.
fare_totals_by_class = [first_class_fare, second_class_fare, third_class_fare]

# 5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.
proportion_by_class = [fare / total_fare for fare in fare_totals_by_class]

# Convert proportions to percentages
percentage_by_class = [round(proportion * 100, 2) for proportion in proportion_by_class]

# Display the results
print("Percentage of total fare revenue by class:")
print(f"First Class: {percentage_by_class[0]}%")
print(f"Second Class: {percentage_by_class[1]}%")
print(f"Third Class: {percentage_by_class[2]}%")
```

```python
# What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?

# 1. Create a new column, age group, that categorizes passengers into "child" (age < 18),
#    "adult" (age 18-64), and "senior" (age 65 and above).
df['age_group'] = pd.cut(df['Age'], bins=[0, 18, 64, float('inf')], labels=['child', 'adult', 'senior'], right=False)

# 2. Calculate the total number of passengers on the Titanic.
total_passengers = len(df)

# 3. Count the number of passengers in each age group.
age_group_counts = df['age_group'].value_counts()

# 4. Divide the count of each age group by the total number of passengers to get the proportion of passengers in each age group.
age_group_proportions = age_group_counts / total_passengers

# 5. Display the proportion as a percentage.
age_group_percentages = (age_group_proportions * 100).round(2)

# Display the results
print("Percentage of passengers by age group:")
print(age_group_percentages)
```
